## 房地产的法律内涵

- 房地产是民法上的物
  > 民法上的物：指存在于人身之外，能够满足人民社会需要而又能为人所实际控制或支配的物质客体
    >> 具有效用
    >> 客观存在
    >> 人身以外
    >> 能被人支配或控制
- 物的分类
  > 流通物与限制流通物、禁止流通物
    >> 流通物：指该物可以在民事主体间自由地流转，自由地转移所有权
         如：食品、国有土地使用权等
    >> 限制流通物：该物只能在特定的民事主体之间流转，受到行政程序的控制
         如：精神性药品、麻醉药品、毒性药品、民用枪支、民用炸药等
    >> 禁止流通物：该物在民事主体之间不能采用民法方法流转
         如：土地所有权，国有所有的土地、矿藏、水流等
  > 原物VS孳息
    >> 原物：指依自然属性或法律的规定，能产生权益的物
    >> 孳息：原物产生之物
      >>> 天然孳息：依自然属性产生的孳息
      >>> 法定孳息：依法律规定产生的孳息
    >> 如母牛生牛犊，房子产生租金，母牛和房子是原物，牛犊和租金是孳息
  > 主物VS从物
    >> 主物：独立存在，与其他独立物结合使用，并在其中发挥主要效用的物
    >> 从物：处于附属地位、起辅助和配合作用的东西
    >> 若没有法律规定，或者合同没有相反约定时，从物的所有权随主物的所有权一并转移
- 动产与不动产
  > 动产：能够移动并且不因移动损害价值的物（金属货币、家具等）
    >> 物大多数属于动产
  > 不动产：不能够移动或虽可移动但却会因移动损害价值的物（房屋等）
    >> 一般以列举的方法界定不动产，而不动产以外的物，都是动产
  > 区分意义：
    >> 所有权人限制：不动产中的土地、河流、森林等只能成为国家或的所有权客体
    >> 设立的他物权类型不同：不动产中的他物权多于动产，物权中的用益物权，如土地使用权、地役权、典权等，只能在不动产上设定；在担保物权中，不动产得设定抵押权，而不能设定质权和留置权，而动产可设定质权和留置权
    >> 不动产发生相邻关系，动产因能移动，不发生特定人之间的相邻关系。民法上有专门处理不动产相邻关系的条文
    >> 地域管辖权不同。民事诉讼法规定：
      >>> 因不动产纠纷提起的诉讼，有不动产所在地人民法院管辖，这叫专属管辖
      >>> 因动产引起的民事诉讼，则依普通管辖确定管辖法院
    >> 物权变动的公示方法不同
      >>> 动产物权的变动——以交付为公示
      >>> 不动产物权的变动——登记为公示，不经登记的不生变动效力

## 什么是物权

### 概念

- 物权是指权利人依法对特定的物享有直接支配和排他的权利
- 民事财产权—特定的物—支配权、绝对权和对世权

### 种类

- 自物权：是对自己所有的物享有的权利（所有权：占有、使用、收益、处分）
- 他物权：对他人的物享有的权利
  > 用益物权：是对他人所有的物享有使用、收益的权利（土地承包经营权、建设用地使用权、宅基地使用权、地役权）
  > 担保物权：担保债务的履行（抵押权、质押权、留置权）
##### 物权的效力
- 物权的排他效力
  > 同一物之上不能同时设立两个所有权
  > 同一物上不能设定两种互相冲突的他物权
  > 物权人有权排除其他一切人对其物权的干涉妨碍
- 物权的优先效力：物权优先与债权
- 物权的追及效力
- 物权的请求效力
  > 确认物权
  > 返还原物
  > 排除妨害
  > 请求清除危险

### 物权本原则

- 物权法定原则
- 一物一物权原则
- 公示、公信原则

## 房地产开发的基本流程

### 房地产开发要办哪些法律手续

- 什么是房地产开发
  > 房地产开发，是指在依法取得国有土地使用权的土地上进行基础设施、房屋建设的行为（《中华人民共和国城市房地产管理法》）；
  > 基础设施——土地一级开发
- 房地产开发属性
  > 固定资产投资；
  > 基本建设
- 房地产开发要办哪些法律手续
  > 拿地—立项—建设用地规划许可证—土地证—建设工程规划许可证—房屋征收—建筑工程施工许可证—商品房预（销售）售许可证—建设工程竣工验收备案

### 房地产开发怎样立项

- 立项类型
  > 审批制：政府投资项目。
  > 核准制：企业投资重大项目和限制类项目的。
  > 备案制：企业投资非重大项目和限制类项目。
- 审批制流程：
  > 项目建议书（发改委）。
  > 规划选址（城乡规划）；用地预审（自然资源）；节能审查（环境保护）；环境影响评价（环境保护）。
  > 可行性研究报告（发改委，附规划选址、用地预审、节能评估和环评审批文件）。
  > 批复。
  > 规划许可手续（城乡规划）；正式用地手续（自然资源）
- 核准制流程
  > 规划选址（城乡规划）；用地预审（自然资源）；节能审查（环境保护）；环境影响评价（环境保护）。
  > 项目申请报告（发改委，附规划选址、用地预审、节能评估和环评审批文件）。
  > 批复。
  > 规划许可手续（城乡规划）；正式用地手续（自然资源）
- 备案制流程
  > 备案手续（发改委）。
  > 批复。
  > 规划选址（划拨用地项目，城乡规划）；用地预审（自然资源）；节能审查（环境保护，普通房地产开发节能登记备案）；环境影响评价（环境保护，普通房地产开发实行？）

## 房地产企业与中介

### 房地产开发企业

- 概念
  > 房地产开发企业是依法律规定的条件和程序设立的，以营利为目的，从事房地产开发和经营的企业；
  > 又叫开发商、发展商、建设单位、房地产开发公司等；
  > 开发商≠投资商
- 设立条件和程序
  > 有符合公司法人登记的名称和组织机构；
  > 有适应房地产开发经营需要的固定的办公用房；
  > 注册资本100万元以上；
  > 有4名以上持有资格证书的房地产专业、建筑工程专业的专职技术人员，2名以上持有资格证书的专职会计人员；
  > 法律、法规规定的其他条件。
  > 设立程序：
    >> “先照后证”；
    >> 《营业执照》（工商局）—备案（房地产开发主管部门）——《暂定资质证书》（房地产开发主管部门）
- 资质
  > 一级；
  > 二级；
  > 三级；
  > 四级。
  > 资质条件：
    >> 完善的质量保证体系；
    >> 商品住宅销售中实行了《住宅质量保证书》和《住宅使用说明书》制度；
    >> 从未发生过重大工程质量事故
- 业务范围
  > 一级；
  > 二级及以下

### 房地产中介

- 房地产中介概念和特点
  > 概念：房地产中介服务是指具有专业执业资格的人员在房地产投资、开发、销售、交易等各个环节中为当事人提供居间服务的经营活动。
  > 特点：
    >> 人员特定；
    >> 委托服务；
    >> 服务有偿
- 中介服务机构及其管理
  > 房地产咨询；
  > 房地产价格评估；
  > 房地产经纪
- 房地产中介服务收费
  > 市场调节价；
  > 不得在标价之外收取未标明的费用

## 房地产开发项目管理规定

### 项目资本金制度

- 房地产开发项目**应当**建立资本金制度，**资本金**占项目总投资比率不得低于20%。—《城市房地产开发经营管理条例》
- 资本金制度房地产开发立项条件？

### 开发迟延项目处理

- 征收土地闲置费
  > 超过土地使用权出让合同约定的动工开发期限1年未动工的，可以征收相当于土地使用权出让金20%以下的土地闲置费；
- 无偿收回
  > 满2年未动工的，可以无偿收回土地使用权
##### 开发项目责任制度
- 房地产开发项目质量责任的承担主体——房地产开发企业
- 勘察、设计、施工、监理等单位应当依照有关法律、法规的规定或者合同的约定，承担相应的责任
- 出现什么质量问题可以退房？
  > 主体结构质量不合格
    >> 商品房—交付使用主体质量不合格—向工程质量监督单位申请重新核验—确属主体结构质量不合格—退房、开发商依法承担赔偿责任
##### 房地产项目转让
- 以出让方式取得的土地使用权
  > 按照出让合同约定已经支付全部土地使用权出让金，并取得土地使用权证书；
  > 按照出让合同约定进行投资开发，属于房屋建设工程的，完成开发投资总额的25%以上，属于成片开发建设的，形成工业用地或其他建设用地条件（投资强度符合要求）
- 以划拨方式取得的土地使用权
  > 必须经有批准权的人民政府审批；；
  > 由受让方先补办土地使用权出让手续，并依照国家有关规定缴纳土地使用权出让金，才能进行转让；
  > 不办理土地使用权出让手续而转让房地产，但转让方应将转让房地产所获收益中的土地收益上缴国家或作其他处理

# 第三章 城乡规划法律制度

- 《中华人民共和国城乡规划法》（2019年修正）

## 城乡规划法律制度

### 城乡规划

- 城乡规划是以促进经济社会全面协调可持续发展为根本任务，促进土地科学使用为基础、促进人居环境根本改变为目的，涵盖城乡居民点的**空间布局规划**（法规效力）
- 简单地讲，城乡规划就是描绘城乡发展蓝图，为城乡发展确定合理布局，综合安排城市各项的工程建设

### 城乡规划和城市体系规划的内容

- 城规规划体系
  > 城镇体系规划；
  > 城市规划；
  > 镇规划；
  > 乡规划；
  > 村规划。
- 城市规划、镇规划
  > 总体规划。
  > 详细规划
    >> 控制性详细规划；
    >> 修建性详细规划。

### 城市规划的内容

- 总体规划
  > 规划区范围；
  > 规划区内建设用地规模；
  > 基础设施和公共服务设施用地；
  > 水源地和水系；
  > 基本农田和绿化用地；
  > 环境保护；
  > 自然与历史文化遗产保护；
  > 防灾减灾等；
- 详细规划

### 城市规划与房地产开发的关系

- 城市规划管理局决定未来房地产开发价值和最有效利用程度
- 城市规划管理局是政府调控地价的手段

### 城乡规划法

《中华人民共和国城乡规划法》（2007年起实施，2019年最新修订）

## 城乡规划的编制与审批

### 城市规划与其他规划的协调

- 国民经济和社会发展规划
- 城市、镇总体规划及村庄规划
- 土地利用总体规划

### 空间规划

- 多规合一（2003年）
  > 将国民经济和社会发展规划、城乡规划、土地利用规划、生态环境保护规划等多个规划融合到一个区域上，实现一个市县一本规划、一张蓝图，解决现有各类规划自成体系、内容冲突、缺乏衔接等问题
- 空间规划（2013 习近平）
  > 更加强调规划的综合性和协调性（一张蓝图绘到底）

### 城镇体系规划的编制与审批

- 全国城镇体系规划
  > 由国务院城乡规划主管部门会同国务院有关部门组织编制，报国务院审批
- 省域城镇体系规划
  > 由省或自治区人民政府编制，报国务院审批
- 城市规划的编制与审批

### 城乡规划的修改

- 经依法批准的城乡规划，是城乡建设和规划管理的依据，未经法定程序不得修改（《中华人民共和国城乡规划法》）
- 可以修改规划的情形
  > 上级人民政府制定的城乡规划发生变更，提出修改规划要求的；
  > 行政区划调整确需修改规划的；
  > 因国务院批准重大建设工程确需修改规划的；
  > 经评估确需修改规划的；
  > 城乡规划的审批机关认为应当修改规划的其他情形。
- 修改要走的法定程序
  > 总结实施情况；
  > 向原审批机关报告；
  > 编制修改方案；
  > 修改；
  > 审批

## 城市规划的实施管理

- 报建审批管理
  > 建设项目选址管理（《项目选址意见书》）；
  > 建设用地规划管理（《建设用地规划许可证》）；
  > 建设工程规划管理（《建设工程规划许可证》）。
- 批后管理
  > 监督检查

### 建设项目选址管理

- 按照国家规定需要有关部门**批准或者核准**的建设项目，以**划拨方式提供国有土地使用权**的，建设单位在报送有关部门批准或者核准前，应当向城乡规划主管部门申请核发**选址意见书**（《中华人民共和国城乡规划法》）
- 选址意见书
  > 由城市规划行政主管部门提出的关于建设项目选在哪个城市或者选在哪个方位的意见等具有法律效力的文件
  > 建设项目的基本情况
  > 建设项目规划选址的依据
  > 建设项目选址、用地范围和具体规划要求

### 建设用地规划管理

- 《建设用地规划许可证》
- 划拨地的建设用地规划管理
  > 《中华人民共和国城乡规划法》第三十七条：在城市、镇规划区内以划拨方式提供国有土地使用权的建设项目，经有关部门批准、核准、备案后，建设单位应当向城市、县人民政府城乡规划主管部门提出建设用地规划许可申请，由城市、县人民政府城乡规划主管部门依据控制性详细规划核定建设用地的位置、面积、允许建设的范围，核发建设用地规划许可证。　　
建设单位在取得建设用地规划许可证后，方可向县级以上地方人民政府土地主管部门申请用地，经县级以上人民政府审批后，由土地主管部门划拨土地。
  > 流程
    >> 选址意见书；
    >> 立项；
    >> 建设用地规划许可申请；
    >> 建设用地规划许可证（未知、范围、面积）；
    >> 申请用地；
    >> 划拨土地。
- 出让地的建设用地规划管理
  > 《中华人民共和国城乡规划法》第三十八条：在城市、镇规划区内以出让方式提供国有土地使用权的，在国有土地使用权出让前，城市、县人民政府城乡规划主管部门应当依据控制性详细规划，提出出让地块的**位置、使用性质、开发强度**等规划条件，作为国有土地使用权出让合同的组成部分。未确定规划条件的地块，不得出让国有土地使用权。　
以出让方式取得国有土地使用权的建设项目，在**签订国有土地使用权出让合同后**，建设单位应当持建设项目的批准、核准、备案文件和国有土地使用权出让合同，向城市、县人民政府城乡规划主管部门**领取建设用地规划许可证**。　　
城市、县人民政府城乡规划主管部门不得在建设用地规划许可证中，擅自改变作为国有土地使用权出让合同组成部分的规划条件。
  > 流程
    >> 规划条件；
    >> 出让；
    >> 签订出让合同；
    >> 立项；
    >> 建设用地规划许可证（位置、开发强度、使用性质）。
- 违反建设用地规划管理的后果
  > **规划条件**未纳入国有土地使用权出让合同的，该国有土地使用权出让合同**无效**；对未取得建设用地规划许可证的建设单位批准用地的，由县级以上人民政府撤销有关批准文件；占用土地的，应当及时退回；给当事人造成损失的，应当依法给予赔偿。

### 建设工程规划管理

- 目的：审查项目的建设工程设计方案，或者修建性详细规划是否符合控制性详细规划和规划条件
- 强制公布
  > 城市、县人民政府城乡规划主管部门或者省、自治区、直辖市人民政府确定的镇政府**应当**依法将经审定的**修建性详细规划、建设工程设计方案的总平面图**予以公布

### 城市规划的实施监督检查

- 开工前的规划验线
- 现场检查
- 竣工验收
  > 未经核实或者经核实不符合规划条件的，建设单位不得组织竣工验收
- 竣工资料的报送
  > 建设单位应当在竣工验收后六个月向城乡规划主管 部门报送有关竣工验收资料
